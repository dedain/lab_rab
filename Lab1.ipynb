### Задание 3
#### Реализовать методы сортировки строк числовой матрицы в соответствии с заданием. Оценить время работы каждого
#### алгоритма сортировки и сравнить его со временем стандартной функции сортировки. Испытания проводить на сгенерированных матрицах.
#### Методы: выбором, вставкой, обменом, Шелла, турнирная, быстрая, пирамидальная

```python
import timeit

first = timeit.default_timer()
#сортировка выбором
def selection(data):
    for i, e in enumerate(data):
        mn = min(range(i, len(data)), key=data.__getitem__)
        data[i], data[mn] = data[mn], e
    return data
data = [-1, 1, 4, 12,24534, 134, 34, 1, 78, 234]  
selection(data)
print (data)


second = timeit.default_timer()
#сортировка вставкой
def insertion_sort(nums):
    for i in range(1, len(nums)):
        temp = nums[i]
        j = i - 1
        while (j >= 0 and temp < nums[j]):
            nums[j + 1] = nums[j]
            j = j - 1
        nums[j + 1] = temp
nums = [-1, 1, 4, 12,24534, 134, 34, 1, 78, 234]
insertion_sort(nums)
print(nums)


third = timeit.default_timer()
#сортировка обменом
def bubbles(array):
    # определяем длину массива
    a = len(array)
    #Внешний цикл, количество проходов N-1
    for i in range(a):
        # Внутренний цикл, N-i-1 проходов
        for j in range(0, a-i-1):
            #Меняем элементы местами
            if array[j] > array[j+1]:
                temp = array[j]
                array[j] = array[j+1]
                array[j+1] = temp
array = [-1, 1, 4, 12,24534, 134, 34, 1, 78, 234]
bubbles(array)
print (array)


fourth = timeit.default_timer()
#сортировка методом Шелла
def shell(num):
    inc = len(num) // 2
    while inc:
        for i, e in enumerate(num):
            while i >= inc and num[i - inc] > e:
                num[i] = num[i - inc]
                i -= inc
            num[i] = e
        inc = 1 if inc == 2 else int(inc * 5.0 / 11)
num = [-1, 1, 4, 12,24534, 134, 34, 1, 78, 234]
shell(num)
print (num)


fifth = timeit.default_timer()
#турнирная сортировка ?????????????????????????????????
def tournament_sort(arr):
    tree = [None] * 2 * (len(arr) + len(arr) % 2)
    index = len(tree) - len(arr) - len(arr) % 2
    
    for i, v in enumerate(arr):
        tree[index + i] = (i, v)
    
    for j in range(len(arr)):
        n = len(arr)
        index = len(tree) - len(arr) - len(arr) % 2
        while index > -1:
            n = (n + 1) // 2
            for i in range(n):
                i = max(index + i * 2, 1) # на последней итерация index + i * 2 = 0
                if tree[i] != None and tree[i + 1] != None:
                    if tree[i][1] < tree[i + 1][1]:
                        tree[i // 2] = tree[i]
                    else:
                        tree[i // 2] = tree[i + 1]
                else:
                    tree[i // 2] = tree[i] if tree[i] != None else tree[i + 1]
            index -= n

        index, x = tree[0]
        arr[j] = x
        tree[len(tree) - len(arr) - len(arr) % 2 + index] = None
arr = [-1, 1, 4, 12,24534, 134, 34, 1, 78, 234]
tournament_sort(arr)
print(arr)


sixth = timeit.default_timer()
#быстрая сортировка
def partition(ar, begin, end):
    a = begin
    for i in range(begin+1, end+1):
        if ar[i] <= ar[begin]:
            a += 1
            ar[i], ar[a] = ar[a], ar[i]
    ar[a], ar[begin] = ar[begin], ar[a]
    return a
def quick_sort(ar, begin=0, end=None):
    if end is None:
        end = len(ar) - 1
    
    def _quicksort(ar, begin, end):
        if begin >= end:
            return
        a = partition(ar, begin, end)
        _quicksort(ar, begin, a-1)
        _quicksort(ar, a+1, end)
    return _quicksort(ar, begin, end)
ar = [-1, 1, 4, 12,24534, 134, 34, 1, 78, 234]
quick_sort(ar)
print(ar)

seventh = timeit.default_timer()
#пирамидальная
# Вспомогательная функция
def heapify(numbers, heap, index):  
    # Индекс наибольшего элемента считается корневым индексом
    larg = index
    left = (2 * index) + 1
    right = (2 * index) + 2

    # Если левый потомок корня — это допустимый индекс, а элемент больше,
    # чем текущий наибольший, то обновляется наибольший элемент
    if left < heap and numbers[left] > numbers[larg]:
        larg = left

    # Для правого потомка корня
    if right < heap and numbers[right] > numbers[larg]:
        larg = right

    # Если наибольший элемент уже не корневой, они меняются местами
    if larg != index:
        numbers[index], numbers[larg] = numbers[larg], numbers[index]
        heapify(numbers, heap, larg)

def heap_sort(numbers):  
    n = len(numbers)

    # Создаём Max Heap из списка
    # 2-й аргумент означает остановку алгоритма перед элементом -1, то есть
    # перед первым элементом списка
    # 3-й аргумент означает повторный проход по списку в обратном направлении, 
    # уменьшая счётчик i на 1 
    for i in range(n, -1, -1):
        heapify(numbers, n, i)

    # Перемещаем корень Max Heap в самый конец списка
    for i in range(n - 1, 0, -1):
        numbers[i], numbers[0] = numbers[0], numbers[i]
        heapify(numbers, i, 0)

# Проверяем, что всё работает
numbers = [-1, 1, 4, 12,24534, 134, 34, 1, 78, 234]
heap_sort(numbers)  
print(numbers)

eighth = timeit.default_timer()
gh = sorted([-1, 1, 4, 12,24534, 134, 34, 1, 78, 234])
print (gh)


print('Стандартная сортировка   ',timeit.default_timer()-eighth) # :^(
print('Метод сортировки выбором ',timeit.default_timer()-first)
print('Метод сортировки вставкой',timeit.default_timer()-second)
print('Метод сортировки обмена  ',timeit.default_timer()-third)
print('Метод сортировки Шелла   ',timeit.default_timer()-fourth)
print('Турнирная сортировка     ',timeit.default_timer()-fifth)
print('Быстрая сортировка       ',timeit.default_timer()-sixth)
print('Пирамидальная сортировка ',timeit.default_timer()-seventh)
```
